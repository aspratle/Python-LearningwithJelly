## Scatterplot


```python
#Let's do a scatterplot of our data

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns #we need the library!

df = pd.read_csv('data/king_county.csv') 
df.head()
```

```python
#simple scatterplot on square feet living and price 

fig, ax = plt.subplots()

X = df['sqft_living']
y = df['price']

sns.scatterplot(x=X, y=y)
```

```python
# can add another column to change the color of the points 
sns.scatterplot(x=X, y=y, hue=df['waterfront'])
```

```python
#change the colors and add labels 

sns.scatterplot(x=X, y=y, hue=df['waterfront'], palette="Set2")
plt.title ("Square Feet Living vs. Price based on Waterfront")
plt.xlabel("Square Feet Living")
plt.ylabel("Price")
plt.show()
```

## Facetted Histograms (Displot)

```python
#multiple histograms at once -- may take a second to run

sns.set_theme(style="darkgrid")

sns.displot(x=df["sqft_living"], col=df["condition"],
           binwidth=3, facet_kws=dict(margin_titles=True),)
```

## Boxplot

```python
sns.set_theme(style="ticks", palette="pastel")

sns.boxplot(x=df["waterfront"], y=df["sqft_living"])
```

## Heatmap

```python
## good for looking at relationships among predictors and predictors vs. target (important in modeling!)

corrs = df.corr()

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corrs, cmap=cmap, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
