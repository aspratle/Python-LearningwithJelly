## Reading in data as a DataFrame 


```python
import numpy as np 
import pandas as pd #need the library!

df = pd.read_csv('data/king_county.csv') #reads in a csv file
type(df)
```

## Explore the Data - Methods & Attributes

```python
# What's in the first 5 rows of my Dataframe? (method) 

df.head(10)
```

```python
# What's in the last 5 rows of my DataFrame? (method)

df.tail()
```

```python
#What are my columns and the datatypes stored in those columns? (method)

df.info()
```

```python
#Summary stats of my columns (method)

df.describe()
```

```python
#How many rows and columns in my dataframe? (attribute)

df.shape
```

```python
#What are my column names? (attribute)

df.columns #may be better to do a .info() 
```

## Locate and filter for certain rows

```python
#loc is for label location filter based on a column label (name)

df.loc[df.bedrooms>=4]
```

```python
#select certain rows I want to look at 

df.iloc[0:15]
```

## I can also get correlations (this is powerful)

```python
df.corr()

# -1 to 1 
```
